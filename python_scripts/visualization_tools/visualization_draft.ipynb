```python
import pandas as pd

def visualize_stock_bars():
    con = get_connection()

    SQL = '''
        SELECT * FROM stock_bars
        '''

    df = pd.read_sql_query(SQL, con)
    print('stock_bars: ')
    print(df.describe())

    print(df.groupby(['symbol']).describe())

visualize_stock_bars()

```
